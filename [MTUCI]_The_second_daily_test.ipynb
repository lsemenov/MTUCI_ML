```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC, SVC
import numpy as np
from sklearn.metrics import accuracy_score, recall_score, precision_score
```

### 6、please do data preprocesing

```python
features = df.drop(["diagnosis"], axis="columns")
target = df['diagnosis']
target=target.astype('int')
```

```python
#необходимо для функции выведения гиперплоскости на графике наблюдения
features= features[["radius_mean","texture_mean"]]
print(features)
```

```python
#стандартизируем признаки
scaler = StandardScaler()
standard_features = scaler.fit_transform(features)

#размечаем данные

X_train, x_test, Y_train, y_test = train_test_split(standard_features, target, test_size = 0.3, random_state=42)
standard_features[:,]
```

### 7、please use SVM to do classification

#### 7.1 LinearSVM

```python
from sklearn.metrics import accuracy_score

#создадим линейный классификатор
svc = LinearSVC(C=1.0)
#тренируем модель
model = svc.fit(X_train, Y_train)

# предсказываем метки класса
predict = model.predict(x_test)
#эффективность модели
print ("Accuracy Score : ",accuracy_score(y_test, predict) )
print ("Recall Score (сколько злокачественных опухолей было предсказано правильно) : ",recall_score(y_test, predict))
print ("Precision Score (сколько опухолей которые были предсказаны, как злокачественные , действительно было злокачественными): ",precision_score(y_test, predict))
```

```python
from matplotlib import pyplot as plt

color = ["black" if c==0 else "lightgrey" for c in y_test]

plt.scatter(x_test[:,0], x_test[:,1], c=color)
w = svc.coef_[0]
a = -w[0]/w[1]
xx = np.linspace(-2.5, 2.5)
yy = a * xx - (svc.intercept_[0])/w[1]

#чертим гиепрплоскоть
plt.plot(xx,yy)
plt.axis("off")
plt.show()
```

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

def plot_decision_regions(X,y, classifier):
    cmap = ListedColormap(("red", "blue"))
    xx1,xx2 = np.meshgrid(np.arange(-3,3, 0.02), np.arange(-3,3,0.02))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.1, cmap=cmap)
    
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y==cl, 0], y=X[y==cl, 1],
                   alpha=0.8, c=cmap(idx),
                   marker="+", label=cl)
```

__Вывод:__ в данном случае классы линейно не сеперабельны, гиперплоскость не может линейно разделить два класса.
В этих случаях существует баланс между максимизацией промежутка гиперплоскости и минимизацией ошибочной классификации. В опорно-векторном классификаторе последнее контролируется с помощью гиперпараметра С — штрафа, налагаемого на ошибки. С — это параметр  SVC и штраф за ошибочное классифицирование точки данных. Когда значение С мало, классификатор получает одобрение на ошибочно классифицированные точки данных (высокое смещение, но низкая дисперсия). Когда значение С большое, классификатор сильно штрафуется за ошибочно классифицированные данные и поэтому отклоняется назад, чтобы избежать любых ошибочно классифицированных точек данных (низкое смещение, но высокая дисперсия). библиотеке scikit-leam значение С определяется параметром с и по умолчанию с= 1.о . Мы должны относиться к С , как к гиперпараметру обучающегося алгоритма
